```python
import pandas as pd
import datetime
from datetime import date,  timedelta
from  pyspark.sql.functions import col,when
from pyspark.sql.types import DoubleType
from pyspark.sql.types import BooleanType
from pyspark.sql.types import IntegerType
```

```python
#afr = sqlContext.sql(""" select * from dnd.pred_afr """)
#afr = afr.toPandas()

apa = sqlContext.sql(""" select * from dnd.pred_apa_v2 """)
apa = apa.toPandas()

eur = sqlContext.sql(""" select * from dnd.pred_eur_prod_v2 """)
eur = eur.toPandas()

imea = sqlContext.sql(""" select * from dnd.pred_imea_prod_v2 """)
imea = imea.toPandas()

lam = sqlContext.sql(""" select * from dnd.pred_lam_prod_v2 """)
lam = lam.toPandas()

nam = sqlContext.sql(""" select * from dnd.pred_nam_prod_v2 """)
nam = nam.toPandas()

#wca = sqlContext.sql(""" select * from dnd.pred_wca """)
#wca = wca.toPandas()

main_data = pd.concat([apa, eur, imea, lam, nam], ignore_index = True)
main_data[['CONTAINER_SIZE_CD', 'Std_FreeTimeGranted', 'TT_pred']] = main_data[['CONTAINER_SIZE_CD', 'Std_FreeTimeGranted', 'TT_pred']].astype(int)
main_data = main_data.sort_values(by=['PCD_Month', 'DIPLA_CITY_CD', 'Spot_YN', 'CONTAINER_SIZE_CD', 'Std_FreeTimeGranted']).reset_index().drop('index', axis=1)
main_data['Cont_Type'] = 'Dry'
main_data = main_data.rename(columns={'PCD_Month': 'Pcd_Month', 'CONSIGNEE_CUSTOMER_CD':'Consignee_CD', 'DIPLA_CITY_CD': 'Dipla_CD', 'CONTAINER_SIZE_CD':'Cont_Size', 'Std_FreeTimeGranted': 'SFT_Granted', 
                                      'TT_pred':'TT_Pred'})

main_data['Operator'] = 'MAEU'
main_data = main_data[['Operator', 'Pcd_Month', 'Dipla_CD', 'Consignee_CD', 'Spot_YN', 'Cont_Type', 'Cont_Size', 'SFT_Granted', 'TT_Pred']]
main_data.tail()
```

```python
Pred_df_main_spark = spark.createDataFrame(main_data)
Pred_df_main_spark.createOrReplaceTempView("Pred_df_main_spark")
#Pred_df_main_spark.write.mode("overwrite").saveAsTable("dnd.Pred_df_main_spark")
Pred_df_main_spark.write.mode("overwrite").saveAsTable("dnd.Pred_df_main_spark_v2")
```
